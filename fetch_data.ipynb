```python
import requests
import requests_html
import pandas as pd
import time
```

```python
links = [
    'https://oto.com.vn/mua-ban-xe-audi/f9605555', # audi 2,3,5,7 cho
    'https://oto.com.vn/mua-ban-xe-ford/f9605555', # ford 2,3,5,7 cho
    'https://oto.com.vn/mua-ban-xe-chevrolet/f9605555', #chevrolet
    'https://oto.com.vn/mua-ban-xe-toyota/f9605555', # toyota
    'https://oto.com.vn/mua-ban-xe-hyundai/f9605555', # huyndai
    'https://oto.com.vn/mua-ban-xe-suzuki/f9605555',
    'https://oto.com.vn/mua-ban-xe-thaco/f9605555',
    'https://oto.com.vn/mua-ban-xe-nissan/f9605555',
]
```

```python
session = requests_html.HTMLSession()
```

```python
for link in links:
    r = session.get(link)
    count = r.html.find('.listcar', first=True).find('.textcount', first=True).find('span', first=True)
    count = count.text
    print(f'{count} : {link}')
```

```python
links = [
    {'type': 'toyota', 'url':'https://oto.com.vn/mua-ban-xe-toyota/f9605555'},
    {'type': 'hyundai', 'url':'https://oto.com.vn/mua-ban-xe-hyundai/f9605555'},
    {'type':'ford', 'url':'https://oto.com.vn/mua-ban-xe-ford/f9605555'},
]
```

```python
def get_cars_on_page(r):
    df = pd.DataFrame(columns=['url'])
    
    list_cars = r.html.find('.listcar', first=True)
    
    for car in list_cars.find('.sellcar-item'):
        post_url = car.find('.photo', first=True).find('a', first=True)
        url = post_url.attrs['href']
        
        df = df.append({'url': url}, ignore_index=True)
    
    return df
```

```python
def get_df_from_link(link):
    df = pd.DataFrame(columns=['url'])

    r = session.get(link)
    df = df.append(get_cars_on_page(r))
    
    
    # get available next pages
    for page_number in range(2, 100):
        
        next_page_link = link+'/p'+str(page_number)
        print(next_page_link)

        r = session.get(next_page_link)
        df = df.append(get_cars_on_page(r))
    return df
    
```

```python
for link in links:
    df = get_df_from_link(link['url'])
    df.to_csv('./data/'+link['type']+'.csv')
    time.sleep(1)
```

# Get photo from links

```python
import os
```

```python
data_dir = './data'
csv_files = os.listdir(data_dir)
csv_files = [csv_file for csv_file in csv_files if os.path.splitext(csv_file)[1] == '.csv']
csv_files
```

```python
df_list = []

for csv_file in csv_files[1:2]:
    filename = os.path.splitext(csv_file)[0]
    file_path = os.path.join(data_dir, csv_file)
    df = pd.read_csv(file_path, index_col=0)
    df = df.assign(type=lambda _: filename)

    df_list.append(df)
    
df = pd.concat(df_list)
df = df.reset_index(drop=True)
df
```

```python
df['full_url'] = df['url'].apply(lambda url: 'https://oto.com.vn' + url)
df
```

```python
df['full_url'].iloc[0]
```

```python
def get_image_urls_from_link(link):
    r = session.get(link)
    
    image_slides = r.html.find('.imageslide', first=True, clean=True)
    if image_slides is None:
        return []
    imgs = image_slides.find('img')

    img_links = [img.attrs['src'] for img in imgs]
    return img_links
```

```python
image_url_df = pd.DataFrame(columns=['type', 'url'])

for i, row in df[['type','full_url']].iterrows():
    url = row['full_url']
    t = row['type']
    
    print(f'{i}/{len(df)}: {url}')
    image_urls = get_image_urls_from_link(url) # list
    image_df = pd.DataFrame(image_urls, columns=['url'])
    image_df = image_df.assign(type=lambda _: t)
    image_url_df = image_url_df.append(image_df)

    time.sleep(0.5)
    
    if (i+1) % 21 == 0:
        image_url_df.to_csv(f'./data/{i}.csv')
```

# Get images

```python
import pandas as pd
import re
import requests
import time
import os
```

```python
df = pd.read_csv('./data/image_urls.csv', index_col=0)
df
```

```python
df = df.reset_index(drop=True)
df
```

```python
df['url'] = df.apply(lambda row: re.sub(r'crop/\d+x\d+/', '', row['url']), axis=1)
```

```python
gdf = df.groupby(['type', 'url'], sort=True, as_index=False).first()
```

```python
df = gdf
```

```python
uniques = df['url'].unique()
len(uniques)
```

```python
df = pd.DataFrame({'url': uniques})
df['type'] = 'huyndai'
df
```

```python
image_dir = './images'
```

```python
if not os.path.exists(image_dir):
    os.mkdir(image_dir)
```

```python
for label in df['type'].unique():
    if not os.path.exists(os.path.join(image_dir, label)):
        os.mkdir(os.path.join(image_dir, label))
```

```python
count = {
    'huyndai': 0
}
```

```python
with open('./progress.txt', 'wt') as progress_file:
    for i in range(8, len(df)):
        label, url = df.iloc[i].type, df.iloc[i].url

        r = requests.get(url)
        if not r.ok:
            print(f'!!!{url}', file=progress_file)
            continue

        img_ext = os.path.splitext(url)[1]
        img_name = count.get(label, 0)
        img_path = os.path.join(image_dir, label, f'{img_name}{img_ext}')

        with open(img_path, 'wb') as f:
            f.write(r.content)
        count[label] = count.get(label, 0) + 1
        print(f'{i}/{len(df)} {img_path}', file=progress_file)

        time.sleep(0.5)
```
